### Terrorist Attacks in a Particular year and their Locations

```python
#year-wise analyis of terrorist attacks
filterYear = df['iyear'] == 1970
filterData = df[filterYear] # filter data
# filterData.info()
reqFilterData = filterData.loc[:,'city':'longitude'] #We are getting the required fields
reqFilterData = reqFilterData.dropna() # drop NaN values in latitude and longitude
reqFilterDataList = reqFilterData.values.tolist()
# reqFilterDataList
```

```python
map = folium.Map(location = [0, 30], tiles='CartoDB positron', zoom_start=2)
# clustered marker
markerCluster = folium.plugins.MarkerCluster().add_to(map)
for point in range(0, len(reqFilterDataList)):
    folium.Marker(location=[reqFilterDataList[point][1],reqFilterDataList[point][2]], popup = reqFilterDataList[point][0]).add_to(markerCluster)
map
```

80% of the terrorist attacks in 1970 were carried out in North and South American. In 1970, the Middle East and North Africa, currently the center of wars and terrorist attacks, faced just one terrorist attack.

### Death Analysis

```python
killData = df.loc[:,'nkill']
print(int(sum(killData.dropna())), "people are killed by terrorists.")
```

```python
attackData = df.loc[:,'attacktype1':'attacktype1_txt']
typeKillData = pd.concat([attackData, killData], axis=1)
typeKillFormatData = typeKillData.pivot_table(columns='attacktype1_txt', values='nkill', aggfunc='sum')
typeKillFormatData
```

```python
labels = typeKillFormatData.columns.tolist() 
transpose = typeKillFormatData.T 
values = transpose['nkill']
fig, ax = plt.subplots(figsize=(10,10), subplot_kw=dict(aspect="equal"))
plt.pie(values, autopct='%.2f%%')
plt.title('Types of terrorist attacks causing deaths',fontsize=15)
plt.legend(labels, loc='upper right', bbox_to_anchor = (1.4, 0.9), fontsize=10) 
plt.show()
```

Armed Assaults and Explosions have caused maximum damage.

```python
countryData = df.loc[:,'country':'country_txt']
countryKillData = pd.concat([countryData, killData], axis=1)
countryKillFormatData = countryKillData.pivot_table(columns='country_txt', values='nkill', aggfunc='sum')
countryKillFormatData
```

```python
transpose = countryKillFormatData.T
transpose.sort_values('nkill', inplace=True, ascending=False)
values = transpose['nkill']
values = [int(i) for i in values]  
values = values[:50]
label = transpose.index
label = label[:50]
index_ = np.arange(len(label))
fig, ax = plt.subplots(figsize=(20,10))
ax.yaxis.grid(True)
plt.bar(index_, values, color = 'red', width = 0.9)
plt.xticks(index_, label, fontsize=15, rotation=90)
plt.yticks(fontsize=15)
plt.title('Countries majorly affected by terrorist attacks', fontsize=15)
plt.show()
```
